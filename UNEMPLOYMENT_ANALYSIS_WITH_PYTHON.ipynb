```python
new_df = df.rename(columns={
    'Region': 'States',
    ' Date': 'Date(yyyy/mm/dd)',
    ' Frequency': 'Frequency(Hz)',
    ' Estimated Unemployment Rate (%)': 'E_UnEmployees_Rate',
    ' Estimated Employed': 'Estimated_Employees',
    ' Estimated Labour Participation Rate (%)': 'Estimated_Labour_Perti_Rate',
    'Region.1': 'Region',
    'longitude': 'Longitude',
    'latitude': 'Latitude'})
```

### Converting 'Date' column to datetime format

```python
import datetime as dt
```

```python
new_df['Date(yyyy/mm/dd)'] = pd.to_datetime(new_df['Date(yyyy/mm/dd)'], errors='coerce').dt.strftime('%Y/%m/%d')
```

### Heatmap Corraletion Matrix

```python
heat_maps = new_df[['E_UnEmployees_Rate','Estimated_Employees', 'Estimated_Labour_Perti_Rate','Longitude', 'Latitude']]

heat_maps = heat_maps.corr()

plt.figure(figsize=(10,6))
sns.set_context('notebook',font_scale=1)
sns.heatmap(heat_maps, annot=True,cmap='Oranges');
```

## There is no strong positive correlation (absolute value greater than 0.7).
## There is a strong negative correlation between E_UnEmployees_Rate and Estimated_Employees with a correlation coefficient of -0.074.

# Data Visualization
creating a Bar Chart to visualize the Estimated Unemployment Rate for different States.

```python
new_df=pd.DataFrame(new_df)
y=new_df['E_UnEmployees_Rate']
x=new_df['States']
pr= plt.figure(figsize=(40, 15))
plt.xlabel("States",fontweight='bold',fontsize=20)
plt.ylabel("Estimated Unemployment Rate(%)",fontweight='bold',fontsize=20)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
plt.yticks(fontweight='bold',fontsize=15)
plt.bar(x,y, color='b',align='center')
plt.show()
```
